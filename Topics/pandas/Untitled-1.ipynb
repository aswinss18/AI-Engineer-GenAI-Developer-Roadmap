```python
import pandas as pd

# Read original CSV
df = pd.read_csv("titanic.csv")

# Rename column
df.rename(columns={"PassengerId": "Id"}, inplace=True)

# Convert 0/1 to False/True strings
df["Survived"] = df["Survived"].replace({
    0: "False",
    1: "True"
})

# Replace all missing values with N/A
df = df.fillna("N/A")

# Output file path
output_file = r"C:\Development\AI-Engineer-GenAI-Developer-Roadmap\Topics\pandas\cleaned\cleaned_titanic.csv"

# Save cleaned CSV
df.to_csv(output_file, index=False,columns=["Id", "Survived", "Name","Age","Sex","Ticket","Fare"])

# Read the SAME file back
cleaned_df = pd.read_csv(
    output_file,
    keep_default_na=False
)

# Display DataFrame
cleaned_df
```
